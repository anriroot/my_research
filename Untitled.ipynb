```python
import requests
from lxml import etree
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import re,sys
from tika import parser
import json
from time import sleep

headers = {"User-agent": "Mozilla/5.0 (Windows NT 6.3; WOW64; rv:50.0) Gecko/20100101 Firefox/50.0"}

mainurl = 'https://egrul.nalog.ru/index.html'

#Наши файлы
MY_EXCEL = 'пример.xlsx'
REZULAT = 'Результат.xlsx'

#Временный пдф название для сохранения
TEMP_PDF = 'temp.pdf'

#пауза между запросами в секундах
MY_TIMEOUT = 1

#Запись в эксель и получение pandas dataframe
def write_excel(list_dicts,name=REZULAT):
    df = pd.DataFrame.from_dict(list_dicts)
    print(df)
    print('Saved to - ' + name + '.xlsx')
    df.to_excel(name + '.xlsx')
    return df

#Считывание из эксель
def read_excel(excel=MY_EXCEL):
    df = pd.read_excel(excel)
    return df.to_dict('records')

#Скачивание пдф
def download_pdf(bot,url):
    r = bot.s.get(url, allow_redirects=True)

    try:
        os.remove(TEMP_PDF)
    except:
        pass

    open(TEMP_PDF, 'wb').write(r.content)

#Получаем текст с пдф
def get_text_from_pdf(pdf_file):
    print('Считываем из файла. Первый запуск может быть долгим. Инициализация')
    raw = parser.from_file(pdf_file)

    return raw['content']

#Берем параметры из пдф
def get_params(txt):
    try:
        email = re.findall('E-mail (.*)',txt)[0]
    except:
        email = 'Не найден'

    try:
        what = re.findall('Сведения об основном виде деятельности(?:\n.*){6}',txt)[0].replace('\n','')
    except:
        what = ''

    return email,what

#Получаем инфу по одному
def getone(info):
    bot = nalogbot()

    inn = info['ИНН']
    print(inn)

    print('Делаем запросы')
    token = bot.search1(inn)

    sleep(MY_TIMEOUT)

    pdf_token = bot.search2(token)

    sleep(MY_TIMEOUT)

    bot.request(pdf_token)
    download_url = 'https://egrul.nalog.ru/vyp-download/' + pdf_token
    sleep(MY_TIMEOUT)

    download_pdf(bot,download_url)
    print('пдф файл сохранен')

    txt = get_text_from_pdf(TEMP_PDF)

    email,what = get_params(txt)
    info['e-mail'] = email
    info['Деятельность'] = what

    return info

#Получаем все значения
def get_all(infos):
    infos2 = []
    for info in infos:
        try:
            info2 = getone(info)
            infos2.append(info2)
            print('Успешно завершен один')
        except Exception as msg:
            print(msg)
            infos2.append(info)

    return infos2

#Класс бота для всех запросов
class nalogbot:

    def __init__(self):
        self.s = requests.session()
        self.headers = headers

    def getmain(self):
        self.s.get(mainurl)

    def search1(self,query):
        headers = {
            'Connection': 'keep-alive',
            'sec-ch-ua': '"Google Chrome";v="95", "Chromium";v="95", ";Not A Brand";v="99"',
            'Accept': 'application/json, text/javascript, */*; q=0.01',
            'Content-Type': 'application/x-www-form-urlencoded; charset=UTF-8',
            'X-Requested-With': 'XMLHttpRequest',
            'sec-ch-ua-mobile': '?0',
            'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36',
            'sec-ch-ua-platform': '"Linux"',
            'Origin': 'https://egrul.nalog.ru',
            'Sec-Fetch-Site': 'same-origin',
            'Sec-Fetch-Mode': 'cors',
            'Sec-Fetch-Dest': 'empty',
            'Referer': 'https://egrul.nalog.ru/index.html',
            'Accept-Language': 'ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7',
        }

        data = {
            'vyp3CaptchaToken': '',
            'page': '',
            'query': query,
            'region': '',
            'PreventChromeAutocomplete': ''
        }

        r = self.s.post('https://egrul.nalog.ru/', headers=headers, data=data)
        return json.loads(r.text)['t']

    def search2(self,token):
        headers = {
            'Connection': 'keep-alive',
            'sec-ch-ua': '"Google Chrome";v="95", "Chromium";v="95", ";Not A Brand";v="99"',
            'Accept': '*/*',
            'X-Requested-With': 'XMLHttpRequest',
            'sec-ch-ua-mobile': '?0',
            'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36',
            'sec-ch-ua-platform': '"Linux"',
            'Sec-Fetch-Site': 'same-origin',
            'Sec-Fetch-Mode': 'cors',
            'Sec-Fetch-Dest': 'empty',
            'Referer': 'https://egrul.nalog.ru/index.html',
            'Accept-Language': 'ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7',
        }


        r = requests.get('https://egrul.nalog.ru/search-result/' + token, headers=headers)
        return json.loads(r.text)['rows'][0]['t']

    def request(self,pdftoken):
        headers = {
            'Connection': 'keep-alive',
            'sec-ch-ua': '"Google Chrome";v="95", "Chromium";v="95", ";Not A Brand";v="99"',
            'Accept': '*/*',
            'X-Requested-With': 'XMLHttpRequest',
            'sec-ch-ua-mobile': '?0',
            'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36',
            'sec-ch-ua-platform': '"Linux"',
            'Sec-Fetch-Site': 'same-origin',
            'Sec-Fetch-Mode': 'cors',
            'Sec-Fetch-Dest': 'empty',
            'Referer': 'https://egrul.nalog.ru/index.html',
            'Accept-Language': 'ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7',
        }

        r = requests.get('https://egrul.nalog.ru/vyp-request/' + pdftoken, headers=headers)
        return json.loads(r.text)

#Основная функция
def main():
    #Считываем наш исходный excel.
    records = read_excel()

    #Получаем новую информацию об email и номерах.
    infos = get_all(records)

    df = write_excel(infos)

    print('Pandas analytics')

    print('Find how many email domains have .RU')
    print(df[df['e-mail'].str.contains(".RU")==True])

    print('Find how many email domains have not .COM')
    print(df[df['e-mail'].str.contains(".COM")!=True])

    print('Find how many inns started from 97')
    print(df.loc[df["ИНН"].astype(str).str.startswith('97')])

    print('Find how many phones have +')
    print(df[df['Телефон'].str.contains("\+")==True])



if __name__ == "__main__":
    main()
```
